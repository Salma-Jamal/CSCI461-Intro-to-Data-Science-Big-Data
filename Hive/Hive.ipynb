### Hive Basic operations:
Here are some operations that Hive can performs:

- Create database<br> ```CREATE DATABASE d```.

- Drop database<br>```DROP DATABASE d```.
    

- Create table<br>
```CREATE TABLE tablename (colname DATATYPE, ...)
ROW FORMAT DELIMITED
FIELDS TERMINATED BY char
LOCATION 'user/cloudera/a';```
- Querying table sql queries e.g. <br>
``` SELECT a FROM  B```
    
- Loading data 
    - ```hdfs dfs -mv /a/b.txt /user/hive/warehouse/b/```
    - ```INSERT INTO TABLE a SELECT * FROM b;```
    
- insert data from local path to table
    - ```LOAD DATA LOCAL INPATH '/home/bitnami/Big_Data_Labs/Lab2/webpage/' INTO TABLE webpage;```
